```python
import cv2
desired_size = 1000
im_pth = "../categorized_images/airplane/000000001761.jpg"
im = cv2.imread(im_pth)
old_size = im.shape[:2] # old_size is in (height, width) format
desired_size = max(old_size)

ratio = float(desired_size)/max(old_size)
new_size = tuple([int(x*ratio) for x in old_size])

# new_size should be in (width, height) format


im = cv2.resize(im, (new_size[1], new_size[0]))

delta_w = desired_size - new_size[1]
delta_h = desired_size - new_size[0]

top, bottom = 0, delta_h
left, right = 0, delta_w


color = [0, 0, 0]
new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,
    value=color)

#cv2.imshow("image", new_im)
cv2.imwrite("test.jpg",new_im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
def padImage(file):# padimage(00000000081.jpg)
    src = "../categorized_images/airplane/"+file
    dst = "../padded_categorized_images/airplane/"+file
    im = cv2.imread(im_pth)
    old_size = im.shape[:2] # old_size is in (height, width) format
    desired_size = max(old_size)

    ratio = float(desired_size)/max(old_size)
    new_size = tuple([int(x*ratio) for x in old_size])

    # new_size should be in (width, height) format


    im = cv2.resize(im, (new_size[1], new_size[0]))

    #get padding sizes
    delta_w = desired_size - new_size[1]
    delta_h = desired_size - new_size[0]

    #paste source in top right of image
    top, bottom = 0, delta_h
    left, right = 0, delta_w


    color = [0, 0, 0]
    new_im = cv2.copyMakeBorder(im, top, bottom, left, right, cv2.BORDER_CONSTANT,
        value=color)

    #cv2.imshow("image", new_im)
    cv2.imwrite(dst,new_im)
```
